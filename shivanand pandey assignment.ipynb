5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a
heatmap.

```python

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Generate synthetic dataset with correlated features
np.random.seed(0)
mean = [0, 0, 0]
cov = [[1, 0.8, 0.7], [0.8, 1, 0.9], [0.7, 0.9, 1]]
data = np.random.multivariate_normal(mean, cov, 1000)

df = pd.DataFrame(data, columns=['Feature1', 'Feature2', 'Feature3'])

# Calculate correlation matrix
corr_matrix = df.corr()

# Visualize correlation matrix using heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', square=True)

# Add title
plt.title('Correlation Matrix')

# Display the plot
plt.show()

```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

# Create DataFrame
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)

# Create 3D surface plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.title('3D Surface Plot')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}
df = pd.DataFrame(data)

# Create bubble chart
plt.scatter(df['GDP'], df['Population'], s=df['Population']*10, alpha=0.5)

# Set labels and title
plt.xlabel('GDP')
plt.ylabel('Population')
plt.title('Bubble Chart: Population vs. GDP')

# Add annotations
for i, row in df.iterrows():
    plt.annotate(row['Country'], (row['GDP'], row['Population']))

# Show the plot
plt.show()
```

#1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

```python

import numpy as np
from bokeh.plotting import figure, show

# Generate x-values from 0 to 10
x = np.linspace(0, 10, 100)

# Calculate y-values as sine of x
y = np.sin(x)

# Create Bokeh plot
p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

# Add line renderer to plot
p.line(x, y, line_width=2)

# Display plot
show(p)
```

#2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.


```python
from bokeh.plotting import figure, output_file, show
import numpy as np
import pandas as pd

# Generate random data
np.random.seed(42)
N = 100
x = np.random.rand(N)
y = np.random.rand(N)
sizes = np.random.randint(10, 100, N)
colors = ["#%02x%02x%02x" % (r, g, b) for r, g, b in zip(np.random.randint(0, 256, N), np.random.randint(0, 256, N), np.random.randint(0, 256, N))]

# Create a DataFrame
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create the figure
p = figure(title="Scatter Plot with Varying Sizes and Colors", x_axis_label='x', y_axis_label='y')

# Plot the data
p.scatter(x='x', y='y', size='sizes', color='colors', source=df)

# Output the plot
output_file("scatter_plot.html")
show(p)
```
